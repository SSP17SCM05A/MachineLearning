```python
#Binarising the features 
```

```python
def Glucose_bi(df):
    d1=df
    Glucose1 = pd.get_dummies(d1['Glucose'], prefix='Glucose')
    d1 = pd.concat([d1,Glucose1], axis=1)
    d1.drop('Glucose', axis=1, inplace=True)
    return d1
```

```python
def DiabetesPedigreeFunction_bi(df):
    for i,rows in df.iterrows():
        if rows['DiabetesPedigreeFunction']>df['DiabetesPedigreeFunction'].mean():
            df['DiabetesPedigreeFunction']=df['DiabetesPedigreeFunction'].replace(rows['DiabetesPedigreeFunction'],1)
        else:
            df['DiabetesPedigreeFunction']=df['DiabetesPedigreeFunction'].replace(rows['DiabetesPedigreeFunction'],0)
            df['DiabetesPedigreeFunction']=df.DiabetesPedigreeFunction.astype(int)
    return df
```

```python
def age_bi(df):
    for i,rows in df.iterrows():
        if rows['Age']>df['Age'].mean():
            df['Age']=df['Age'].replace(rows['Age'],1)
        else:
            df['Age']=df['Age'].replace(rows['Age'],0)
            df['Age']=df.Age.astype(int)
    return df    
```

```python
def BMI_bi(df):
    for i,rows in df.iterrows():
        if rows['BMI']>df['BMI'].mean():
            df['BMI']=df['BMI'].replace(rows['BMI'],1)
        else:
            df['BMI']=df['BMI'].replace(rows['BMI'],0)
    df['BMI']=df.BMI.astype(int)
    return df
```

```python
def Pregnancies_bi(df):
    d1= df
    Pregnancies1 = pd.get_dummies(d1['Pregnancies'], prefix='Pregnancies')
    d1 = pd.concat([d1,Pregnancies1], axis=1)
    d1.drop('Pregnancies', axis=1, inplace=True)
    return d1
```

```python
def BloodPressure_bi(df):
    d1= df
    BloodPressure1 = pd.get_dummies(d1['BloodPressure'], prefix='BloodPressure')
    d1 = pd.concat([d1,BloodPressure1], axis=1)
    d1.drop('BloodPressure', axis=1, inplace=True)
    return d1
```

```python
def Insulin_bi(df):
    d1= df
    Insulin1 = pd.get_dummies(d1['Insulin'], prefix='Insulin')
    d1 = pd.concat([d1,Insulin1], axis=1)
    d1.drop('Insulin', axis=1, inplace=True)
    return d1
```

```python
def SkinThickness_bi(df):
    d1= df
    SkinThickness1 = pd.get_dummies(d1['SkinThickness'], prefix='SkinThickness')
    d1 = pd.concat([d1,SkinThickness1], axis=1)
    d1.drop('SkinThickness', axis=1, inplace=True)
    return d1
```

```python
df=Glucose_bi(df)
df=DiabetesPedigreeFunction_bi(df)
df=age_bi(df)
df=BMI_bi(df)
df=Pregnancies_bi(df)
df=BloodPressure_bi(df)
df=Insulin_bi(df)
df=SkinThickness_bi(df)
```
